```python
data_f = pd.read_csv('src/fracs/data.csv')
data_f.head()
```

```python
L = data_f['L'].to_list()
L_log = np.log10(L)
cs = data_f['cluster_size'].to_list()
cs_log = np.log10(cs)

xln =  np.linspace(0, 5, 10)

z = np.polyfit(L_log, cs_log, 1)
z
```

```python
pz = np.poly1d(z)
yln = pz(xln)
```

```python
# plot line chart
plt.plot(L_log, cs_log, 'o')
plt.plot(xln, yln)

plt.title('Polyfit chart')
plt.xlabel('p')
plt.ylabel('s')

plt.show()
```
